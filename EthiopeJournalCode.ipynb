```python
#Plot linear regression of f(LST) = NDVI
LST = df['LST'].values
NDVI = df['NDVI'].values

# Calculate linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(NDVI, LST)

# Calculate regression line values
regression_line = slope * NDVI + intercept

# Plot the data and regression line
plt.scatter(NDVI, LST, label='Data')
plt.plot(NDVI, regression_line, color='red', label='Linear Regression')

# Add labels and legend
plt.xlabel('NDVI')
plt.ylabel('LST')
plt.legend()

plt.show()

# Print the regression results
print("Slope (m):", slope)
print("Intercept (b):", intercept)
print("R-squared (R^2):", r_value**2)
print("P-value:", p_value)
print("Standard Error:", std_err)
```

# Section 3

```python
LST = df['LST'].values
NDVI = df['NDVI'].values
NDWI = df['NDWI'].values
BU = df['BU'].values
NDSI = df['NDSI'].values

# Create a matrix of independent variables
X = np.column_stack((NDVI, NDWI, BU, NDSI))

# Add a constant (intercept) term to the independent variables
X = sm.add_constant(X)

# Fit the multiple linear regression model
model = sm.OLS(LST, X).fit()

# Get predicted LST values from the model
predicted_LST = model.predict(X)

# Plot the actual vs. predicted LST values
plt.figure(figsize=(10, 6))
plt.scatter(LST, predicted_LST)
plt.xlabel('Actual LST')
plt.ylabel('Predicted LST')
plt.title('Actual vs. Predicted LST')
#plt.grid(True)

# Add a diagonal line for reference
plt.plot([min(LST), max(LST)], [min(LST), max(LST)], color='red', linestyle='--')

# Save the plot as a PNG file after it's displayed
plt.savefig('Actual vs. Predicted LST.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()

# Print R-squared and parameter statistics
print("R-squared:", model.rsquared)
print("\nParameter Statistics:")
print(model.summary())
```

```python
# Calculate the Mean Squared Error (MSE)
mse = mean_squared_error(LST, predicted_LST)

# Print the MSE
print("Mean Squared Error (MSE):", mse)
```
